```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path().resolve().parent.parent))

from defense.src.features import (
    extract_domains, get_tld, is_leetspeak, is_typosquatting,
    extract_features, FEATURE_NAMES,
)
from defense.src import config
```

## استخراج الدومينات


```python
tests = [
    'ادخل على https://stc.com.sa/login',
    'زر www.mobily.com.sa للمزيد',
    'https://www.bankalbilad.com.golv/',
    'http://bit.ly/abc123',
    'تم تفعيل باقتك بنجاح',
]
for msg in tests:
    print(f'  {msg[:50]:50s} → {extract_domains(msg)}')
```

## كشف Typosquatting و Leetspeak


```python
tests = ['zain.com.sa','za1n.com.sa','bankalbilad.com','bankalbilad.com.golv','mobily.cc']
print(f'{"Domain":<25s} {"Leet":<8s} {"Typo":<8s}')
print('-'*45)
for d in tests:
    print(f'  {d:<25s} {str(is_leetspeak(d)):<8s} {str(is_typosquatting(d)):<8s}')
```

## الخصائص الـ 39 على رسالة واحدة


```python
msg = 'تحذير! فزت بجائزة من موبايلي! اضغط على mobily-rewards.cc لاستلامها فوراً'
f = extract_features(msg)
print(f'Message: {msg}\n')
for fname in FEATURE_NAMES:
    print(f'  {fname:<20s} {f[fname]}')
```

## مقارنة: تصيد vs آمنة


```python
phishing = extract_features('تحذير! فزت بجائزة. اضغط على stc-rewards.cc')
safe = extract_features('تم تفعيل باقتك الشهرية. شكراً لاختيارك زين')

keys = ['has_url','sus_tld','brand_imp','trusted_dom','urgency','reward','threat','action','risk']
print(f'{"Feature":<18s} {"Phishing":<12s} {"Safe":<12s}')
print('-'*45)
for k in keys:
    print(f'  {k:<18s} {phishing[k]:<12} {safe[k]:<12}')
```
