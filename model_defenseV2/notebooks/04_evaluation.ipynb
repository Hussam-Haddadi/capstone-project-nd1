```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path().resolve().parent.parent))

import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix, f1_score, recall_score, precision_score, accuracy_score
from model_defenseV2.src.detector import PhishingDetector

detector = PhishingDetector('../models/defence_model_v2.pkl')
print(detector)
```

```python
df = pd.read_csv('../../data/processed/final_dataset_v2.csv')
results = detector.predict_batch(df['text'].tolist())

y_true = (df['Label'] == 0).astype(int).values
y_pred = [int(r['is_phishing']) for r in results]

cm = confusion_matrix(y_true, y_pred)
print(f'=== Confusion Matrix ===')
print(f'  TN={cm[0][0]:<6d} FP={cm[0][1]:<6d}')
print(f'  FN={cm[1][0]:<6d} TP={cm[1][1]:<6d}')
print(f'\n=== Metrics ===')
print(f'  Accuracy:  {accuracy_score(y_true, y_pred):.4f}')
print(f'  Recall:    {recall_score(y_true, y_pred):.4f}')
print(f'  Precision: {precision_score(y_true, y_pred):.4f}')
print(f'  F1:        {f1_score(y_true, y_pred):.4f}')
```

## تحليل الأخطاء


```python
df['pred'] = y_pred
df['actual'] = y_true

fn = df[(df['actual']==1) & (df['pred']==0)]
fp = df[(df['actual']==0) & (df['pred']==1)]

print(f'False Negatives (missed phishing): {len(fn)}')
for _, row in fn.head(5).iterrows():
    print(f'  {row["text"][:100]}')

print(f'\nFalse Positives (safe flagged): {len(fp)}')
for _, row in fp.head(5).iterrows():
    print(f'  {row["text"][:100]}')
```
