```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path().resolve().parent.parent))

import pandas as pd
import numpy as np
import re
from collections import Counter

df = pd.read_csv('../../data/processed/final_dataset_v2.csv')
print(f'Dataset: {len(df)} rows')
print(f'  Phishing (0): {(df["Label"]==0).sum()}')
print(f'  Safe (1):     {(df["Label"]==1).sum()}')
print(f'  Ratio:        {(df["Label"]==0).sum()/(df["Label"]==1).sum():.2f}:1')
```

## جودة البيانات


```python
print(f'Nulls:      {df.isnull().sum().sum()}')
print(f'Duplicates: {df.duplicated(subset="text").sum()}')

# Length stats
for label, name in [(0,'Phishing'),(1,'Safe')]:
    lengths = df[df['Label']==label]['text'].str.len()
    print(f'{name:10s}: min={lengths.min():.0f}  median={lengths.median():.0f}  mean={lengths.mean():.0f}  max={lengths.max():.0f}')
```

## تحليل الروابط والأرقام


```python
from model_defenseV2.src.features import extract_domains

for label, name in [(0,'Phishing'),(1,'Safe')]:
    subset = df[df['Label']==label]
    with_url = subset['text'].apply(lambda t: len(extract_domains(t)) > 0).sum()
    with_phone = subset['text'].str.contains(r'05\d{8}|920\d{6}', regex=True).sum()
    print(f'{name}: URLs={with_url}/{len(subset)} ({with_url/len(subset)*100:.0f}%) | Phones={with_phone}')
```
