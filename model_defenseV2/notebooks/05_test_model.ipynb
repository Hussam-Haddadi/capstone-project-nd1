````python
import sys
from pathlib import Path
sys.path.insert(0, str(Path().resolve().parent.parent))

from model_defenseV2.src.detector import PhishingDetector
detector = PhishingDetector('../models/defence_model_v2.pkl')
print(f'✅ Model loaded: {detector}')
````

## ✏️ اختبر رسالة واحدة


````python
message = 'زاجل: شحنتك في الطريق. الموعد المتوقع: الأحد. علشان تتابع شحنتك، اضغط هنا: www.mobily.com.cc'

r = detector.predict(message)
icon = '🔴' if r['is_phishing'] else '🟢'
print(f'{icon} {r["label_ar"]}')
print(f'   Risk: {r["risk_score"]:.1f} | Confidence: {r["confidence"]} | Method: {r["detection_method"]}')
for flag in r['flags']:
    print(f'   • {flag}')
````

## اختبار عدة رسائل


````python
messages = [
    'تحذير! تم اكتشاف تسجيل دخول. حدّث بياناتك على bank-secure.online',
    'مكتب خدمات استقدام للتواصل 0501234567',
    'ولي العهد وجه بمشروع جديد http://bit.ly/2JBe1lG',
    'تم تفعيل باقتك الشهرية بنجاح. شكراً لاختيارك زين',
    'الراجحي: عملية شراء 500 ريال. للاستفسار: 920003344',
    'مرحبا نبيل رئيس مشروع نيوم: ولي العهد وجه بإعداد لوائح ملائمة للاستثمار في المشروع http://bit.ly/2JBe1lG',
]

for msg in messages:
    r = detector.predict(msg)
    icon = '🔴' if r['is_phishing'] else '🟢'
    print(f'{icon} {r["label_ar"]:15s} | risk={r["risk_score"]:5.1f} | {msg[:65]}')
````
