# 03 · Generator Training

تدريب الـ N-gram language model على أنماط الرسائل الأصلية.

---


```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path().resolve().parent / 'src'))
sys.path.insert(0, str(Path().resolve().parent.parent))

import pandas as pd
from generator import PhishingGenerator

df = pd.read_csv('../../data/raw/sms_dataset.csv').drop_duplicates(subset='text')
phish_texts = df[df['Label']==0]['text'].tolist()
print(f'Training corpus: {len(phish_texts)} phishing messages')
```

## تدريب الـ N-gram Model


```python
gen = PhishingGenerator(seed=42)
gen.fit(phish_texts)
print('✓ Generator fitted!')
print(f'  N-gram starters: {len(gen.ngram.starters)}')
print(f'  Word transitions: {len(gen.ngram.word_model)}')
```

## اختبار التوليد


```python
categories = ['brand','reward','delivery','gov','phone','leet','shortener','social','safe']
for cat in categories:
    msgs = gen.generate(category=cat, count=2)
    icon = '🔴' if cat != 'safe' else '🟢'
    print(f'\n{icon} {cat.upper()}')
    for m in msgs:
        print(f'  {m["text"][:90]}...')
```

## حفظ الـ Generator


```python
gen.save('../models/attack_generator.pkl')
print('✓ Saved: attack_generator.pkl')
```
