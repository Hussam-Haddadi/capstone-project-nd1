# 01 · Data Cleaning

تنظيف البيانات الخام `sms_dataset.csv` وتجهيزها للاستخدام.

---


```python
import pandas as pd
import numpy as np
import re

df = pd.read_csv('../../data/raw/sms_dataset.csv')
print(f'Raw data: {len(df)} rows × {len(df.columns)} columns')
print(f'Columns: {list(df.columns)}')
df.head()
```

## فحص المشاكل


```python
print('=== Quality Check ===')
print(f'Nulls: {df.isnull().sum().sum()}')
print(f'Duplicates: {df.duplicated(subset="text").sum()}')
print(f'Empty texts: {(df["text"].str.strip() == "").sum()}')
print(f'Very short (<10 chars): {(df["text"].str.len() < 10).sum()}')

# Label conflicts
conflicts = df.groupby('text')['Label'].nunique()
print(f'Conflicting labels: {(conflicts > 1).sum()}')

print(f'\n=== Label Distribution ===')
print(f'Phishing (0): {(df["Label"]==0).sum()}')
print(f'Safe (1):     {(df["Label"]==1).sum()}')
```

## التنظيف


```python
# حذف التكرارات
df_clean = df.drop_duplicates(subset='text').reset_index(drop=True)
print(f'After removing duplicates: {len(df_clean)} rows (removed {len(df) - len(df_clean)})')

# حذف الرسائل الفارغة أو القصيرة جداً
df_clean = df_clean[df_clean['text'].str.strip().str.len() >= 10].reset_index(drop=True)
print(f'After removing short messages: {len(df_clean)} rows')

print(f'\nFinal clean data:')
print(f'  Phishing: {(df_clean["Label"]==0).sum()}')
print(f'  Safe:     {(df_clean["Label"]==1).sum()}')
print(f'  Total:    {len(df_clean)}')
```
