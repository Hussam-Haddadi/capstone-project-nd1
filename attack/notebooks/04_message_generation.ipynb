```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path().resolve().parent / 'src'))
sys.path.insert(0, str(Path().resolve().parent.parent))

import pandas as pd
from generator import PhishingGenerator
import config as atk_config
```

```python
df_orig = pd.read_csv('../../data/raw/sms_dataset.csv').drop_duplicates(subset='text')
orig_phishing = df_orig[df_orig['Label']==0]['text'].tolist()
orig_safe = df_orig[df_orig['Label']==1]['text'].tolist()

gen = PhishingGenerator(seed=42)
gen.fit(orig_phishing)
print(f'Original: {len(orig_phishing)} phishing + {len(orig_safe)} safe')
```

## توليد الداتا سيت


```python
dataset = gen.generate_dataset(
    n_phishing=800,
    n_safe=600,
    original_phishing=orig_phishing,
    original_safe=orig_safe,
)

df_gen = pd.DataFrame(dataset)
print(f'Generated: {len(df_gen)} total')
print(f'  Phishing: {(df_gen["label"]==0).sum()}')
print(f'  Safe:     {(df_gen["label"]==1).sum()}')
print(f'  Ratio:    {(df_gen["label"]==0).sum()/(df_gen["label"]==1).sum():.2f}:1')
```

```python
# Category distribution
print('=== Phishing Categories ===')
for cat, cnt in df_gen[df_gen['label']==0]['category'].value_counts().items():
    pct = cnt / (df_gen['label']==0).sum() * 100
    bar = '█' * int(pct / 2)
    print(f'  {cat:12s} {bar} {cnt} ({pct:.1f}%)')
```
