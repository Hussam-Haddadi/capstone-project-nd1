# 05 · Build Final Dataset

تجميع كل الرسائل في `final_dataset.csv` — الناتج النهائي لموديل الهجوم.

هذا الملف هو اللي يتدرب عليه موديل الدفاع.

---


```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path().resolve().parent / 'src'))
sys.path.insert(0, str(Path().resolve().parent.parent))

import pandas as pd
import numpy as np
from generator import PhishingGenerator
```

```python
df_orig = pd.read_csv('../../data/raw/sms_dataset.csv').drop_duplicates(subset='text')
orig_phishing = df_orig[df_orig['Label']==0]['text'].tolist()
orig_safe = df_orig[df_orig['Label']==1]['text'].tolist()

gen = PhishingGenerator(seed=42)
gen.fit(orig_phishing)
dataset = gen.generate_dataset(n_phishing=800, n_safe=600, original_phishing=orig_phishing, original_safe=orig_safe)
```

## التحقق من الجودة


```python
df = pd.DataFrame(dataset)
print('=== Quality Check ===')
print(f'Total:      {len(df)}')
print(f'Phishing:   {(df["label"]==0).sum()}')
print(f'Safe:       {(df["label"]==1).sum()}')
print(f'Ratio:      {(df["label"]==0).sum()/(df["label"]==1).sum():.2f}:1')
print(f'Duplicates: {df.duplicated(subset="text").sum()}')
print(f'Nulls:      {df.isnull().sum().sum()}')
```

## حفظ الناتج النهائي


```python
df_export = df[['text', 'label']].rename(columns={'label': 'Label'})
df_export = df_export.sample(frac=1, random_state=42).reset_index(drop=True)
df_export.to_csv('../../data/processed/final_dataset.csv', index=False)
print(f'✅ Saved: data/processed/final_dataset.csv ({len(df_export)} rows)')
print(f'\nThis dataset feeds into the defense model training.')
```
